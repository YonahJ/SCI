```python
import pandas as pd
from collections import Counter
import os
import numpy as np
import pandas as pd
from os import path
from PIL import Image
import codecs
from gensim.models.word2vec import Word2Vec
import matplotlib.pyplot as plt
```

```python
# This function organizes corpus as list of sentences, each sentence as a list of words,
## as input to the function WordVec

## Collect preprocessed texts in txt format
#root_folder = "C:/Users/test/Dropbox/Baumard/Frontiers/scripts/3. Generate historical semantic map/preprocessed_texts/"
root_folder = os.getcwd() + '\preprocessed_texts'
print(root_folder)
word2vec_input = []

for path, subdirs, files in os.walk(root_folder):
    for file in files:
        if '.txt' in file:
            print(file)
            name = os.path.join(path, file)
        
            file_text = open(name, encoding = 'utf-8').read()
            text_list_paragraphs = file_text.split('\n')

            for paragraph in text_list_paragraphs:
                #print (paragraph)
                
                paragraph = paragraph.replace('\r', '')
                paragraph_list = []
                for word in paragraph.split(' '):
                    #print (word)
                    
                    paragraph_list += [word]
                    
                word2vec_input += [paragraph_list]
            
                #print(word2vec_input)
```

```python
## Here we build the vector space with Word2Vec

SentenceCorpus = word2vec_input
word2vec_output = Word2Vec(SentenceCorpus, min_count=1)
```

```python
## Save vector space

word2vec_output.save('w2v_model.txt')
```
